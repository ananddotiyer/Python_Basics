#### Problem6

Consider the following recursive program. How many times is the function **power** called when the above code is executed? In general, can you edit the above code so that it counts the number of times **power** is called?

```python
def power(x, n):
    '''Return x raised to the power n'''
    if n == 0:
        return 1
    if n % 2 == 0:
        root = power(x, n // 2)
        return root * root
    else:
        return x * power(x, n - 1)
power(2, 4)
```
